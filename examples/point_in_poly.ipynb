```python
import numpy as np
```

```python
from holey.point import Point
from holey.loop import winding_number
import matplotlib.pyplot as plt
import holey.draw
```

```python
%matplotlib inline
```

```python
# TODO: Reconcile loop definition as array vs as points (numba knowledge required).
loop_pt = [Point(0, 0), Point(2, 7), Point(6, 6)]
loop = np.array([[0, 0], [2, 7], [6, 6]])
point = Point(3, 4)
```

```python
ax = plt.axes()
holey.draw.loop(ax, loop_pt)
holey.draw.point(ax, point)
contains = winding_number(loop, point) != 0
plt.title(f'Point in loop: {contains}')
```

```python
point = Point(3, 3)

ax = plt.axes()
holey.draw.loop(ax, loop_pt)
holey.draw.point(ax, point)
contains = winding_number(loop, point) != 0
plt.title(f'Point in loop: {contains}')
```
