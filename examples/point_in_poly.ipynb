```python
import numpy as np
```

```python
from holey.point import Point
from holey.polygon import winding_number
import matplotlib.pyplot as plt
import holey.draw
```

```python
%matplotlib inline
```

```python
# TODO: Reconcile poly definition as array vs as points (numba knowledge required).
poly_pt = [Point(0, 0), Point(2, 7), Point(6, 6)]
poly = np.array([[0, 0], [2, 7], [6, 6]])
point = Point(3, 4)
```

```python
ax = plt.axes()
holey.draw.polygon(ax, poly_pt)
holey.draw.point(ax, point)
contains = winding_number(poly, point) != 0
plt.title(f'Point in poly: {contains}')
```

```python
point = Point(3, 3)

ax = plt.axes()
holey.draw.polygon(ax, poly_pt)
holey.draw.point(ax, point)
contains = winding_number(poly, point) != 0
plt.title(f'Point in poly: {contains}')
```
